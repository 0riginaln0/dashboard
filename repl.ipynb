```python
import sqlite3

def run_sql_file(db_path, sql_file_path):
    """
    Execute all SQL statements in a file against a SQLite database.
    
    Parameters:
        db_path (str): Path to the SQLite database file.
        sql_file_path (str): Path to the .sql file to execute.
    """
    conn = sqlite3.connect(db_path)
    with open(sql_file_path, 'r', encoding='utf-8') as f:
        sql_script = f.read()
    conn.executescript(sql_script)
    conn.commit()
    conn.close()
    print(f"Executed {sql_file_path} on {db_path}")
```

```python
DB_NAME = "foo.db"
```

```python
run_sql_file(DB_NAME, "migrations/up.sql")
```

```python

run_sql_file(DB_NAME, "migrations/down.sql")
```

```python
import aiosqlite

async def get_connection(db_path: str):
    conn = await aiosqlite.connect(db_path)
    await conn.execute("PRAGMA journal_mode=WAL")
    return conn
```

```python
import asyncio
import aiosql
import aiosqlite

queries = aiosql.from_path("./sql/queries.sql", "aiosqlite")
print(queries)
```

```python


async def get_users():
    async with aiosqlite.connect("foo.db") as conn:
        users = [user async for user in queries.get_users(conn)]
        print(users)
await get_users()
```

```python
async def add_user():
    async with aiosqlite.connect("foo.db") as conn:
        res = await queries.add_user(conn, name="Aboba", age=21)
        await conn.commit()
        print(type(res))
        print(res)
await add_user()
```

```python
async def find_user():
    async with aiosqlite.connect("foo.db") as conn:
        user = await queries.find_user(conn, id=3)
        print(user)
await find_user()
```

```python
async def delete_user():
    async with aiosqlite.connect("foo.db") as conn:
        user = await queries.delete_user(conn, id=3)
        await conn.commit()
        print(user)
await delete_user()
```
